```python
import math
import cmath
```

```python
def cRound(y):
    return complex(round(y.real,3),round(y.imag,3))
#complex round function
n = int(input('Bitte variable n festlegen:'))
for k in range (0,n):
    a = complex(0,2*cmath.pi*k)
    x = cmath.exp((a)/n)
    print(cRound(x))
    print(cmath.polar(cRound(x)))
    differenz=float(1-cmath.polar(x)[0])
    if abs(differenz)>0.001:
        print('der Punkt liegt nicht auf dem Einheitskreis!')
#Root of unity, polar/cartesian coordiantes
```
